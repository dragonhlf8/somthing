```

#@title **Create VPS-RDP**
#@markdown Enter Username and Password

import os


username = "GranMC" #@param {type:"string"}
password = "root" #@param {type:"string"}

print(f"Creating '{username}' user")

# Creation of user
os.system(f"useradd -m {username}")

# Add user to sudo group
os.system(f"adduser {username} sudo")

# Set password of user to 'root'
os.system(f"echo '{username}:{password}' | sudo chpasswd")

# Change default shell from sh to bash
os.system("sed -i 's/\/bin\/sh/\/bin\/bash/g' /etc/passwd")

print(f"'{username}' created")

#@title **Rdp**
#@markdown  It takes 4-5 minutes for installation

import os
import subprocess

#@markdown  Visit [remote headless](http://remotedesktop.google.com/headless) and Copy the command after authentication

CRP = "DISPLAY= /opt/google/chrome-remote-desktop/start-host --code=\"4/0AX4XfWiJ6VFtlY_0fXoMRCbehf4K48lpoaLKBotQqZckArqqqMqoOIjDqgzsyQdwdAMAgw\" --redirect-url=\"https://remotedesktop.google.com/_/oauthredirect\" --name=$(hostname)" #@param {type:"string"}

#@markdown Enter a pin more or equal to 6 digits
Pin =  123456#@param {type: "integer"}

#@markdown **Select browser** that you want to install. YOU CAN SELECT ALL
GoogleChrome = True #@param {type:'boolean'}
Onion = False #@param {type:'boolean'}
FireFox = False #@param {type:'boolean'}



class CRD:
    def __init__(self):
        os.system("apt update")
        self.installCRD()
        self.installDesktopEnvironment()
        self.runOS()
        self.installAPPS()
        self.finish()

    @staticmethod
    def installCRD():
        print("Installing Chrome Remote Desktop")
        subprocess.run(['wget', 'https://dl.google.com/linux/direct/chrome-remote-desktop_current_amd64.deb'], stdout=subprocess.PIPE)
        subprocess.run(['dpkg', '--install', 'chrome-remote-desktop_current_amd64.deb'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', '--assume-yes', '--fix-broken'], stdout=subprocess.PIPE)

    @staticmethod
    def installDesktopEnvironment():
        print("Installing Desktop Environment")
        os.system("export DEBIAN_FRONTEND=noninteractive")
        os.system("apt install --assume-yes xfce4 desktop-base xfce4-terminal")
        os.system("bash -c 'echo \"exec /etc/X11/Xsession /usr/bin/xfce4-session\" > /etc/chrome-remote-desktop-session'")
        os.system("apt remove --assume-yes gnome-terminal")
        os.system("apt install --assume-yes xscreensaver")
        os.system("systemctl disable lightdm.service")

    @staticmethod
    def runOS():
        if GoogleChrome:
            print("installing GoogleChrome")
            subprocess.run(["wget", "https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb"], stdout=subprocess.PIPE)
            subprocess.run(["dpkg", "--install", "google-chrome-stable_current_amd64.deb"], stdout=subprocess.PIPE)
            subprocess.run(['apt', 'install', '--assume-yes', '--fix-broken'], stdout=subprocess.PIPE)
        if Onion:
            print("installing Onion browser")
            os.system("add-apt-repository ppa:micahflee/ppa")
            os.system("apt update")
            os.system("apt install torbrowser-launcher")
        if FireFox:
            print("installing Firefox browser")
            os.system("apt install firefox")

    @staticmethod
    def installAPPS():
        print("Installing obs studio ")
        subprocess.run(['apt', 'install', 'ffmpeg'], stdout=subprocess.PIPE)
        subprocess.run(['apt',  'update'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', 'v4l2loopback-dkms'], stdout=subprocess.PIPE)
        subprocess.run(['add-apt-repository', 'ppa:obsproject/obs-studio'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'update'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', 'obs-studio'], stdout=subprocess.PIPE)
        subprocess.run(['wget', 'https://github.com/Grannnnnnnnx/RESPI/releases/download/1.10/images.jpeg'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', 'ufw'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', 'snapd'], stdout=subprocess.PIPE)
        print("installing vlc")
        subprocess.run(['apt', 'install', 'vlc'], stdout=subprocess.PIPE)
        subprocess.run(['dd', 'if=/dev/zero', 'of=/mnt/swap.0' 'bs=5024', 'count=1048576'], stdout=subprocess.PIPE)
        subprocess.run(['mkswap', '/mnt/swap.0'], stdout=subprocess.PIPE)
        subprocess.run(['su'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'update'], stdout=subprocess.PIPE)
        subprocess.run(['echo', '"/mnt/swap.0', 'swap', 'swap', 'defaults', '0', '0"', '>>', '/etc/fstab'], stdout=subprocess.PIPE)
        subprocess.run(['swapon', '/mnt/swap.0'], stdout=subprocess.PIPE)
        subprocess.run(['swapon', '-s'], stdout=subprocess.PIPE)


    @staticmethod
    def finish():
        print("Finalizing")
        os.system(f"adduser {username} chrome-remote-desktop")
        command = f"{CRP} --pin={Pin}"
        os.system(f"su - {username} -c '{command}'")
        os.system("service chrome-remote-desktop start")
        print("done")


try:
    if username:
        if CRP == "":
            print("Please enter authcode from the given link")
        elif len(str(Pin)) < 6:
            print("Enter a pin more or equal to 6 digits")
        else:
            CRD()
except NameError as e:
    print("username variable not found")
    print("Create a User First")
```
